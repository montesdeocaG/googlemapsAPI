B. Create a function to Find the Place Id given Place Name (String) and Location (Lat/Lont Tuple).

```python
import requests
import json
def idfinder(strname,locationtuple,radius):
    ###Description: 
        #this funciton gets the ID based on string location and radius
    ###Parameters:
        #strname: string name of the place
        #locationtuple: lat,lon tuple
        #radius: radius seach will be performed in meters
    ###Returns:
        #a place_id object
    data = json.loads(requests.get('https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={},{}&radius={}&key={}'.format(locationtuple[0],locationtuple[1],radius,'APIKEYGOESHERE')).content)
    df1 = pd.DataFrame(data['results'])
    ID = df1[df1['name'] == strname]['place_id']
    return ID
idfinder('Chalmuch',(20.988459,-89.736768),2000)
```
